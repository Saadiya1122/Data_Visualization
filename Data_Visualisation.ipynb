    import re
    def remove_char(text):
        if isinstance(text, str):
            text = re.sub('[^a-zA-Z0-9(/)]',' ', text)
            text = re.sub('\s+',' ', text) #remove whitespace
            return text
        else:
            return text

### Number of Movies and TV Shows<a href="#Number-of-Movies-and-TV-Shows" class="anchor-link">¶</a>

The below code helps provide an overview of the distribution of Movies
and TV shows in the dataset by visually representing the count and
percentage of each content type in a bar chart. The code starts by
taking the count of the 'type' column and keeping it in 'count_by_type'
variable using value_counts() function to count the occurrences of each
content type of Movie and TV shows and then the total count is then
calculated by summing up the counts and then taking the percentage of
the count by dividing the counts(no. of occurences of each content) by
the sum of it and then multiply it by 100. Next to visualize it we use
plotly and express the graph in bar chart, the x-axis and y-axis
represents the content type and the count of movies and TV shows
respectively. The plot is given a title of "Distribution of Movies and
TV Shows". Additionally, labels are added to the bars to display the
count and percentage of each content type. The update_traces() function
is used to add text labels (count_labels) to the bars, with the count
and percentage values formatted appropriately. Finally, the plot is
displayed using fig.show().

The output shows that there are around 70% (to be accurate 69.69%)
Movies and around 30% (30.31%) TV Shows on Netflix's platform.

In \[20\]:

    # Calculate the count or percentage of movies and TV shows
    count_by_type = df['type'].value_counts()
    total_count = count_by_type.sum()
    percentage_by_type = count_by_type / total_count * 100

    # Create a bar plot to visualize the count or percentage
    fig = px.bar(
        x=count_by_type.index,
        y=count_by_type.values,
        labels={'x': 'Type', 'y': 'Count'},
        title='Distribution of Movies and TV Shows'

    # Add labels on the bars with count and percentage
    count_labels = [f'{count} ({percentage:.2f}%)' for count, percentage in zip(count_by_type.values, percentage_by_type.values)]
    fig.update_traces(text=count_labels, textposition='auto')

    fig.show()

### Country-wise Content - the distribution of content across different countries.<a
href="#Country-wise-Content----the-distribution-of-content-across-different-countries."
class="anchor-link">¶</a>

With the below code we will get to know about the country-wise content
on Netflix. Code starts by by counting the titles of each country and
for this we have instantiate a variable that keeps the value counts and
indexes of the 'country' column and then seperates the two column into
'Country' and 'Count'. Followed by loading the geospatial data for
countries from the 'gapminder' dataset included in Plotly
(px.data.gapminder()). The data for the year 2007 is selected, and only
the 'iso_alpha' (ISO 3166-1 alpha-3 country codes) and 'country' columns
are kept. Then we the line merges the geo_data and country_counts based
on the 'country' column in geo_data and the 'Country' column in
country_counts. The merge is performed using a left join, which ensures
all rows from geo_data are included in the merged dataframe. Then we
will generate a choropleth map using Plotly Express (px.choropleth()).
The merged_data dataframe is used as the input data. The 'iso_alpha'
column provides the country codes for the locations on the map, then the
'Count' column determines the color intensity for each country, and the
'country' column is displayed hovering over a country. Then we will set
the title of the map and the color scale is set to as 'Blues'. Then next
we will customize the layout of the figure then set the geo parameter to
control the display of frames, coastlines, and the projection type. In
this case, frames and coastlines are hidden, and the projection type is
set to 'equirectangular'. The annotations parameter adds an annotation
to the figure at the specified position (x=0.5, y=-0.1) with the text
'Number of Titles'.

The output shows that most of the content are from 'USA' with count of
2809 contents on Netflix's platform and 'India' with a count of 972
contents on the platform.

In \[25\]:

    # Count the number of titles per country
    country_counts = df['country'].value_counts().reset_index()
    country_counts.columns = ['Country', 'Count']

    # Load the geospatial data for countries
    # Replace 'world' with 'usa' for a map of the United States
    geo_data = px.data.gapminder().query("year==2007")[['iso_alpha', 'country']]

    # Merge the counts with the geospatial data
    merged_data = pd.merge(geo_data, country_counts, left_on='country', right_on='Country', how='left')

    # Generate the choropleth map
    fig = px.choropleth(merged_data, locations='iso_alpha', color='Count',
                        hover_name='country', title='Distribution of Content Across Countries',
                        color_continuous_scale='Blues')

    # Customize the labels
    fig.update_layout(geo=dict(showframe=False, showcoastlines=False, projection_type='equirectangular'),
                      annotations=[dict(x=0.5, y=-0.1, showarrow=False, text='Number of Titles', font=dict(size=12))])

    fig.show()

# Conclusion<a href="#Conclusion" class="anchor-link">¶</a>

In conclusion, our analysis of the Netflix dataset has uncovered
fascinating insights into the platform's content landscape. Here's a
summary of the key findings:

#### 1 - Distribution of Films and Television Series:<a href="#1---Distribution-of-Films-and-Television-Series:"
class="anchor-link">¶</a>

Movies make up approximately 70% of Netflix's content, while TV shows
constitute the remaining 30%. This emphasizes the platform's vast
collection of movies available for streaming.

#### 2 - Well-known Sorts:<a href="#2---Well-known-Sorts:" class="anchor-link">¶</a>

Among film genres, documentaries hold a prominent position, followed
closely by "Dramas International Movies." This indicates a strong
viewership interest in these genres and their popularity among Netflix
subscribers.

#### 3 - Trends by Release Year:<a href="#3---Trends-by-Release-Year:" class="anchor-link">¶</a>

Over time, the number of movie releases on Netflix has seen a
significant increase compared to TV shows. Particularly after the year
2000, the gap between movies and TV show releases has widened, with a
decline in the number of TV show releases.

#### 4 - Analyses of Ratings:<a href="#4---Analyses-of-Ratings:" class="anchor-link">¶</a>

"TV-MA" emerges as the most common rating for Netflix content,
indicating a substantial portion of the available content is intended
for mature audiences. This suggests a diverse range of content catering
to different age groups.

#### 5 - Content by country:<a href="#5---Content-by-country:" class="anchor-link">¶</a>

The majority of content on Netflix originates from the United States,
followed closely by India. This highlights the significant presence and
contribution of these countries in content creation or acquisition.

The most common duration for Netflix movies is around 90 minutes. In the

#### 7 - Genre Distribution by Age Group:<a href="#7---Genre-Distribution-by-Age-Group:"
